```python
import pylab
import imageio
filename = '../data/develop.mp4'
```

```python
import sys
import skvideo.io
cap = skvideo.io.VideoCapture(filename)
ret, frame = cap.read()
```

```python

vid = imageio.get_reader(filename,  'ffmpeg')
nums = [10, 2000]
for num in nums:
    image = vid.get_data(num)
    fig = pylab.figure()
    fig.suptitle('image #{}'.format(num), fontsize=20)
    pylab.imshow(image)
pylab.show()
```
